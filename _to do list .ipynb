```python
import json
import os

DATA_FILE = "todo_data.json"

# Load existing to-dos
def fetch_data():
    if not os.path.exists(DATA_FILE):
        return []
    with open(DATA_FILE, 'r') as f:
        return json.load(f)

# Save current to-dos
def store_data(todo_list):
    with open(DATA_FILE, 'w') as f:
        json.dump(todo_list, f, indent=2)

# Display the current to-do list
def list_tasks(todo_list):
    if not todo_list:
        print("\nðŸ“‚ No tasks yet.")
        return
    print("\nðŸ“Œ To-Do Tasks:")
    for idx, item in enumerate(todo_list, 1):
        symbol = "âœ”" if item['done'] else "âœ–"
        print(f"{idx}. [{symbol}] {item['task']}")

# Add a new task
def create_task(todo_list):
    task = input("âž• Enter a task: ").strip()
    if task:
        todo_list.append({'task': task, 'done': False})
        print("âœ… Task added.")
    else:
        print("âš  Task cannot be empty.")

# Mark task as done
def finish_task(todo_list):
    list_tasks(todo_list)
    try:
        number = int(input("â˜‘ Task number to mark as done: "))
        if 1 <= number <= len(todo_list):
            todo_list[number - 1]['done'] = True
            print("ðŸ‘ Task marked as done.")
        else:
            print("â— Invalid task number.")
    except ValueError:
        print("âš  Please enter a number.")

# Remove a task
def remove_task(todo_list):
    list_tasks(todo_list)
    try:
        number = int(input("ðŸ—‘ Task number to remove: "))
        if 1 <= number <= len(todo_list):
            removed = todo_list.pop(number - 1)
            print(f"ðŸ§¹ Removed: {removed['task']}")
        else:
            print("â— Invalid task number.")
    except ValueError:
        print("âš  Please enter a number.")

# Main application loop
def todo_main():
    todos = fetch_data()
    while True:
        print("\n===== TO-DO MANAGER =====")
        print("1ï¸âƒ£  View Tasks")
        print("2ï¸âƒ£  Add Task")
        print("3ï¸âƒ£  Mark as Completed")
        print("4ï¸âƒ£  Delete Task")
        print("5ï¸âƒ£  Exit")
        choice = input("Pick an option (1-5): ")

        actions = {
            '1': lambda: list_tasks(todos),
            '2': lambda: create_task(todos),
            '3': lambda: finish_task(todos),
            '4': lambda: remove_task(todos),
            '5': lambda: "exit"
        }

        if choice == '5':
            store_data(todos)
            print("ðŸ‘‹ Exiting... Tasks saved.")
            break
        elif choice in actions:
            actions[choice]()
        else:
            print("âŒ Invalid option. Try again.")

if __name__ == "__main__":
    todo_main()
```
